```python
import pickle
import matplotlib.pyplot as plt
import random
```

```python
with open('output_data/zs20.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
data_dict = {}
for i in data:
    if i[1] not in data_dict:
        data_dict[i[1]] = [i[0]]
    else:
        data_dict[i[1]].append(i[0])

for i in range(9):
    A, B, C = [], [], []
    for k, v in data_dict.items():
        c = int(k)
        for z in v:
            a = z[i]
            b = z[i+1]
            A.append(a)
            B.append(b)
            C.append(c)

    plt.figure(figsize=(10,10))
    plt.scatter(A, B, c=C, cmap='brg', s=5)
    plt.colorbar()
    plt.show()
```
